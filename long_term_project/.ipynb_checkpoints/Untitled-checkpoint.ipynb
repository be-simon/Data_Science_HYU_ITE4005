```python
import pandas as pd
import numpy as np
```

```python
columns = ['user_id', 'item_id', 'rating', 'time_stamp']
u1_base = pd.read_csv('./data/u1.base', sep='\t', names=columns)
u1_test = pd.read_csv('./data/u1.test', sep='\t', names=columns)
```

```python
# get sparse matrix
user_rating = u1_base.drop('time_stamp', axis=1)
user_item_rating = user_rating.pivot_table('rating',index='user_id', columns='item_id').fillna(0)
user_item_rating
```

### item-based collaborative filltering
* 아이템간 유사도 측정
* user - item 의 rating을 측정한다
    * user가 rating 한 item들을 유사도 순으로 나열
    * 해당 rating들을 평균 내서 결정
    

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine similarity
item_cs_array = cosine_similarity(user_item_rating.T)
```

```python
# item correlation
item_index = user_item_rating.columns
item_cs_df = pd.DataFrame(item_cs_array, index=item_index, columns=item_index)
item_cs_df
```

```python
def expect_user_item_rating(user_id, item_id):
    user_rating = user_item_rating.loc[user_id]
    item_corr = item_cs_df[item_id]
    
    rating_corr = pd.DataFrame([user_rating, item_corr], index=['rating', 'correlation']).T
    top_rating = rating_corr[rating_corr['rating'] > 0].sort_values('correlation', ascending=False).head(10)
    
    values = top_rating['rating'].values
    weight = top_rating['correlation'].values
    
    return np.dot(values, weight) / sum(weight)
```

```python
user_rating = user_item_rating.loc[1]
item_corr = item_cs_df[6]

rating_corr = pd.DataFrame([user_rating, item_corr], index=['rating', 'correlation']).T
top_rating = rating_corr[rating_corr['rating'] > 0].sort_values('correlation', ascending=False).head(10)
top_rating['rating'].values
```

```python
user_item_pair = u1_test.drop(['time_stamp', 'rating'], axis=1)
user_item_pair
```

```python
result = []
for idx, user_id, item_id in user_item_pair.itertuples():
    if item_id in user_item_rating.columns:
        result.append(expect_user_item_rating(user_id, item_id))
    else:
        result.append(0)
```

```python
user_item_pair['rating'] = result
user_item_pair
```

```python
original = u1_test['rating'].values
original
```

```python
sum = 0
for p, o in zip(result, original):
    sum += pow(p - o, 2)
```
